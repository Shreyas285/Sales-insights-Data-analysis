# Sales insights Data analysis

Atliq corporation is a computer hardware company which supplies computer hardware and peripherals to its clients across India. Having its headquarters situated in Delhi, it also has its branches situated at various locations within the country. The company is facing issues in terms of tracking its sales, order and revenue inflow at various branches in dynamically growing market. This company is also facing challenges in obtaining insights from the sales data for decision support.

As a data analyst, I have been tasked from Sales Director of the company to create a powerful Power BI dashboard to track past and recent trends in the sales and provide better insights about the business for data driven decision making.

## Project planning

**Purpose**:
To untap sales insight from the data which was not visible before in it's tabular form for decision making and automate the process to reduce time spent on data gathering manually.

**Results** : 
To create a automated Power BI dashboard for providing quick and quality sales insights in order to support data driven decision making.

**Success criteria** : 
* Dashboard uncovering sales, orders and revenue insights with latest data available.
* Sales team able to take data driven better decisions regarding sales.
* Sales Analysts stop manual data gathering process in order to save their business time and re-invest it in value added process.

### Data munging

#### ETL (Extract, Transform and Load) using Power BI

ETL is a process that extracts the data from different source systems, then transforms the data (like applying calculations, concatenations, etc.) and finally loads the data into the central repository.

MySQL database is integrated to Power BI acting as a data source. All the tables from the database are extracted into Power BI. This process is known as extraction.

In the next step the extracted data is transformed using Power Query. The data is cleaned, standardized, normalized and corrected using the tool to make it analysis ready. This process is known as data transform.

In the last step the the transformed data is loaded into Power BI to create a Dashboard and to do further analysis.

#### Data Modeling

Data models are made up of entities, which are the objects or concepts we want to track data about, and in this project they are the tables in a database. Data modeling is a process used to connect multiple entities using a relationship. The relationship is achieved by mapping attributes that are common for different tables.


In this project Star schema is employed for Data modeling. The transaction table is made as fact table and remaining tables as dimensional table.


After achieving the relationship between tables we will proceed to create Power BI Dashboard.

### Creating Dashboard using Power BI

**Problem** :

The dashboard includes,
* Two slicers namely Year and Month to track trends based on years and months.
